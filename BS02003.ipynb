```python
import FlowCal
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = FlowCal.io.FCSData('\\Users\Sigur\Desktop\FCS_Test_03\BS02003_germ_neg.fcs')
s = FlowCal.transform.to_rfi(s)
```

```python
fig = plt.figure()
FlowCal.plot.hist1d(s, channel='FL2-H', facecolor='#440154', title='BS02003 Negative Control 1')
plt.xlim([0, 1000000])
plt.ylim([0, 5000])
import os
plt.show()
fig.savefig('\\Users\Sigur\Desktop\FCS_Test_03\Histogram_BS02003.pdf')
```

```python
fig = plt.figure()
FlowCal.plot.density2d(s, channels=['FSC-H', 'FL2-H'], mode='scatter', cmap='viridis', title='BS02003 Negative Control 1')
plt.xlim([5000, 1000000])
plt.ylim([-1000, 1000000])
import os
plt.show()
fig.savefig('\\Users\Sigur\Desktop\FCS_Test_03\Scatter_BS02003.png', dpi=500)
```
