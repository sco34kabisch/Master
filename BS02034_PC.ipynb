```python
import FlowCal
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = FlowCal.io.FCSData('\\Users\Sigur\Desktop\FCS_Test_34\BS02034_pos.fcs')
s = FlowCal.transform.to_rfi(s)
```

```python
fig = plt.figure()
FlowCal.plot.hist1d(s, channel='FL2-H', facecolor='#fde725', title='BS02034 Positive Control')
plt.xlim([0, 1000000])
plt.ylim([0, 5000])
import os
plt.show()
fig.savefig('\\Users\Sigur\Desktop\FCS_Test_34\Histogram_BS02034_2.pdf')
```

```python
fig = plt.figure()
FlowCal.plot.density2d(s, channels=['FSC-H', 'FL2-H'], mode='scatter', cmap='viridis', title='BS02034 Positive Control')
plt.xlim([5000, 1000000])
plt.ylim([-1000, 1000000])
import os
plt.show()
fig.savefig('\\Users\Sigur\Desktop\FCS_Test_34\Scatter_BS02034.png', dpi=500)
```
