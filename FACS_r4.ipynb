```python
import FlowCal
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = FlowCal.io.FCSData('\\Users\Sigur\Desktop\FCS_Test_r4\S34012_r4.fcs')
s = FlowCal.transform.to_rfi(s)
```

```python
FlowCal.plot.density2d(s, channels=['FSC-A', 'FL2-A'], mode='scatter', cmap='viridis')
plt.xlim([1000, 1000000])
plt.ylim([-1000, 1000000])
plt.show()
```

```python
fig = plt.figure()
FlowCal.plot.hist1d(s, channel='FL2-H', facecolor='#1f9a8a', title='S34012_r4')
plt.xlim([0, 1000000])
plt.ylim([0, 5000])
import os
plt.show()
fig.savefig('\\Users\Sigur\Desktop\FCS_Test_r4\Histogram_S34012_r4.pdf')
```

```python
fig = plt.figure()
FlowCal.plot.density2d(s, channels=['FSC-H', 'FL2-H'], mode='scatter', cmap='viridis', title='S34012_r4')
plt.xlim([5000, 1000000])
plt.ylim([-1000, 1000000])
import os
plt.show()
fig.savefig('\\Users\Sigur\Desktop\FCS_Test_r4\Scatter_S34012_r4.png', dpi=500)
```
