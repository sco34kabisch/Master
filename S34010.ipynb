```python
import FlowCal
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf as pdf_backend
```

```python
s = FlowCal.io.FCSData('\\Users\Sigur\Desktop\FCS_Test_10\S34010_neg.fcs')
s = FlowCal.transform.to_rfi(s)
```

```python
fig = plt.figure()
FlowCal.plot.hist1d(s, channel='FL2-H', facecolor='#481467', title='S34010 Negative Control 2')
plt.xlim([0, 1000000])
plt.ylim([0, 5000])
plt.show()
import os
fig.savefig('\\Users\Sigur\Desktop\FCS_Test_10\Histogram_S34010_2.pdf')
```
