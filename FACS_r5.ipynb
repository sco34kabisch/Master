```python
import FlowCal
import numpy as np
import matplotlib.pyplot as plt
s = FlowCal.io.FCSData('\\Users\Sigur\Desktop\FCS_Test_r5\S34012_r5.fcs')
s = FlowCal.transform.to_rfi(s)
```

```python
fig = plt.figure()
FlowCal.plot.density2d(s, channels=['FSC-H', 'FL2-H'], mode='scatter', cmap='viridis', title='S34012_r5')
plt.xlim([5000, 1000000])
plt.ylim([-1000, 1000000])
import os
plt.show()
fig.savefig('\\Users\Sigur\Desktop\FCS_Test_r5\Scatter_S34012_r5.png', dpi=500)
```

```python
fig = plt.figure()
FlowCal.plot.hist1d(s, channel='FL2-H', facecolor='#24868e', title='S34012_r5')
plt.xlim([0, 1000000])
plt.ylim([0, 5000])
import os
plt.show()
fig.savefig('\\Users\Sigur\Desktop\FCS_Test_r5\Histogram_S34012_r5.pdf')
```
